```python
import pandas as pd
from scipy.stats import zscore

# Load the dataframe
df = pd.read_csv('/content/eeg_data.csv')


# Step 1: Get skewness of each numeric column
skewness = df.select_dtypes(include='number').skew()

# Step 2: Define thresholds
# You can adjust threshold: |skew| > 1 = skewed, <= 1 = not skewed
skewed_cols = skewness[abs(skewness) > 1].index.tolist()
non_skewed_cols = skewness[abs(skewness) <= 1].index.tolist()

# Optional: print count
print(f"Skewed columns: {len(skewed_cols)}")
print(f"Non-skewed columns: {len(non_skewed_cols)}")

def clean_outliers(df, skewed_cols, non_skewed_cols, z_thresh=3, iqr_thresh=1.5):
    df_cleaned = df.copy()
    initial_rows = len(df_cleaned)

    # Handle skewed columns with IQR
    for col in skewed_cols:
        Q1 = df_cleaned[col].quantile(0.25)
        Q3 = df_cleaned[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - iqr_thresh * IQR
        upper_bound = Q3 + iqr_thresh * IQR
        df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]
        if df_cleaned.empty:
            print(f"Warning: Cleaning column '{col}' removed all rows. Returning original DataFrame.")
            return df

    # Handle non-skewed columns with Z-score
    for col in non_skewed_cols:
        # Check if the column is numeric before applying zscore
        if pd.api.types.is_numeric_dtype(df_cleaned[col]):
            z_scores = zscore(df_cleaned[col])
            df_cleaned = df_cleaned[abs(z_scores) < z_thresh]
            if df_cleaned.empty:
                print(f"Warning: Cleaning column '{col}' removed all rows. Returning original DataFrame.")
                return df

    rows_removed = initial_rows - len(df_cleaned)
    print(f"Number of rows removed during outlier cleaning: {rows_removed}")
    return df_cleaned
df_cleaned = clean_outliers(df, skewed_cols, non_skewed_cols)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.preprocessing import StandardScaler

def compare_models_with_plot(X, y):
    models = {
        "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
        "Logistic Regression": LogisticRegression(max_iter=1000),
        "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss')
    }

    cv = StratifiedKFold(n_splits=5)
    results = []

    for name, model in models.items():
        scores = cross_validate(model, X, y, cv=cv, scoring=['accuracy', 'roc_auc', 'f1'])
        results.append({
            'Model': name,
            'Accuracy': scores['test_accuracy'].mean(),
            'ROC-AUC': scores['test_roc_auc'].mean(),
            'F1-score': scores['test_f1'].mean()
        })

    # Convert to DataFrame
    df_results = pd.DataFrame(results)
    print("\n📊 Model Comparison:\n", df_results)

    # Plotting
    df_results.set_index('Model')[['Accuracy', 'ROC-AUC', 'F1-score']].plot(kind='bar', figsize=(10,6))
    plt.title("Model Performance Comparison")
    plt.ylabel("Score")
    plt.ylim(0, 1)
    plt.xticks(rotation=0)
    plt.grid(axis='y')
    plt.legend(loc='lower right')
    plt.tight_layout()
    plt.show()

    return df_results
# Assume df_cleaned is your cleaned DataFrame
X_selected = df_cleaned[selected_features]
y = df_cleaned['epileptic_']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_selected)

# Compare and plot
results_df = compare_models_with_plot(X_scaled, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Assume df_cleaned is your cleaned DataFrame
# Separate features (X) and target (y)
X = df_cleaned.drop('epileptic_', axis=1)
y = df_cleaned['epileptic_']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 1: Define parameter grid to search
param_grid = {
    'n_estimators': [200, 400],        # Number of trees in the forest
    'max_depth': [None, 10, 20],       # Max depth of each tree
    'min_samples_split': [2, 5],       # Minimum samples required to split an internal node
    'min_samples_leaf': [1, 2]         # Minimum samples required in a leaf node
}

# Step 2: Create base model
rf = RandomForestClassifier(random_state=42, n_jobs=-1)

# Step 3: Grid search setup (5-fold CV, optimize F1)
grid = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    scoring='f1',
    cv=StratifiedKFold(n_splits=5),
    n_jobs=-1,
    verbose=1
)

# Step 4: Fit grid search to your data
grid.fit(X_scaled, y)

# Step 5: Print the best parameters and F1 score
print("Best Random Forest Params:", grid.best_params_)
print("Best Random Forest F1     :", grid.best_score_)
```

```python
best_rf = grid.best_estimator_
```

```python
from sklearn.calibration import CalibratedClassifierCV

# Calibrate the trained model
calibrated_rf = CalibratedClassifierCV(best_rf, method='sigmoid', cv=5)


calibrated_rf.fit(X_scaled, y)
```

```python
from sklearn.calibration import calibration_curve
import matplotlib.pyplot as plt

y_prob = calibrated_rf.predict_proba(X_scaled)[:, 1]
prob_true, prob_pred = calibration_curve(y, y_prob, n_bins=10)

plt.plot(prob_pred, prob_true, marker='o', label='Calibrated RF')
plt.plot([0, 1], [0, 1], linestyle='--', label='Perfect Calibration')
plt.xlabel("Predicted Probability")
plt.ylabel("True Probability")
plt.legend()
plt.title("Calibration Curve")
plt.grid()
plt.show()
```

```python
# Before calibration
probs_uncalibrated = best_rf.predict_proba(X_scaled)[:, 1]
prob_true1, prob_pred1 = calibration_curve(y, probs_uncalibrated, n_bins=10)

# After calibration
probs_calibrated = calibrated_rf.predict_proba(X_scaled)[:, 1]
prob_true2, prob_pred2 = calibration_curve(y, probs_calibrated, n_bins=10)

# Plot both
plt.plot(prob_pred1, prob_true1, label='Uncalibrated', marker='o')
plt.plot(prob_pred2, prob_true2, label='Calibrated', marker='o')
plt.plot([0, 1], [0, 1], '--', label='Perfect Calibration')
plt.xlabel("Predicted Probability")
plt.ylabel("True Probability")
plt.legend()
plt.title("Calibration Curve Comparison")
plt.grid()
plt.show()

```
